```python
import mogger
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import datasets
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
def massive_upload(models, model, dataset_train, dataset_test, X_train, X_test, y_train, y_test, param_distributions, n_prev_mod):
    for i, mod in enumerate(models):
        # random params
        for param, val in param_distributions.items():
            par = np.random.choice(val, 1)[0]
            model.set_params(**{param: par})
            
        model.fit(X_train, y_train)
        
        pred = model.predict(X_test)
        
        print(mogger.upload_model(model, dataset_train, mod, 'target', 'desc', ['tag1', 'tag2']))
        print("Model", mod, "is uploaded.\n")
        
        acc = metrics.accuracy_score(y_test, pred)
        auc = metrics.roc_auc_score(y_test, pred)
        r2 = metrics.r2_score(y_test, pred)
        f1 = metrics.f1_score(y_test, pred)
        
        print(mogger.upload_audit(i+1+n_prev_mod, dataset_test, 'acc', acc))
        print("test acc for", mod, "is uploaded.\n")
        print(mogger.upload_audit(i+1+n_prev_mod, dataset_test, 'auc', auc))
        print("test auc for", mod, "is uploaded.\n")
        print(mogger.upload_audit(i+1+n_prev_mod, dataset_test, 'r2', r2))
        print("test r2 for", mod, "is uploaded.\n")
        print(mogger.upload_audit(i+1+n_prev_mod, dataset_test, 'f1', f1))
        print("test f2 for", mod, "is uploaded.\n")
        
        
        pred = model.predict(X_train)
        
        acc = metrics.accuracy_score(y_train, pred)
        auc = metrics.roc_auc_score(y_train, pred)
        r2 = metrics.r2_score(y_train, pred)
        f1 = metrics.f1_score(y_train, pred)
        
        print(mogger.upload_audit(i+1+n_prev_mod, dataset_train, 'acc', acc))
        print("test acc for", mod, "is uploaded.\n")
        print(mogger.upload_audit(i+1+n_prev_mod, dataset_train, 'auc', auc))
        print("test auc for", mod, "is uploaded.\n")
        print(mogger.upload_audit(i+1+n_prev_mod, dataset_train, 'r2', r2))
        print("test r2 for", mod, "is uploaded.\n")
        print(mogger.upload_audit(i+1+n_prev_mod, dataset_train, 'f1', f1))
        print("test f2 for", mod, "is uploaded.\n")
        
```

```python
massive_upload(
    ["H2oGBM", "LM", "NNet", "RandomForest", "DecisionTree", "SVM"],
    SVC(), 1, 2, X_train, X_test, y_train, y_test,
    {
        'C': 2.0**np.arange(-10, 11, 1),
        'gamma': 2.0**np.arange(-10, 11, 1)
    }, 0
)
```
